## Lambda Functions <br>
<p>Lambda functions... or "Anonymous Functions" are referring to inline functions with no name. The keyword lambda denotes the no name function, and executes within a single line. Without saving it to a variable; however, it is not able to be used, unless passed in either as a paramater or within list comprehension.<br>Written as "(keyword lambda) (one or more inputs) (colon) (function to be executed)"</p>

```python
def print_name(name):
    return f"Hello, {name}"

print_name("Ripal")
```

```python
b_print_name = lambda name:f"Hello, {name}"
```

```python
b_print_name("Eli")
```

```python
def print_other_name(name): return f"Hello, {name}"

print_other_name("Adam")
```

```python
num_list = [12, 23, 34, 45, 56]
new_list = []

for n in num_list:
    num = n**2
    if n%2!=0:
        new_list.append(n**2)
    
print(new_list)
```

```python
# one-liner list comprehension
[{ 'id': 0, 'value': n**2 } for n in num_list if n%2!=0]
```

```python
suits = ['diamonds', 'hearts', 'clubs', 'spades']
ranks = range(1, 14)

[(r, s) for s in suits for r in ranks][:5]
```

```python
cards = []
for s in suits:
    for r in ranks:
        cards.append((r, s))
print(cards)
```

```python
# filter(f, a_list)
my_list = [1, 2, 3, 7, 16]

def my_func(num):
    if num < 4:
        return True
    
f = lambda x:x+10 if x%2==0 else None
```

```python
list(filter(my_func, my_list))
```

```python
# The map function return a value for every iteration of a list. Can change values on a map function
list(map(f, my_list))
```

```python
# The filter function only returns the value that evaluate to True. Are NOT able to change values 
# on a filter function
list(filter(f, my_list))
```

```python
from statistics import mean

nums = [2, 7, 4.2, 1.6, 9, 4.4, 4.9]

# Aggregate Function
# pulling a bunch of values within an array into a singular value
# max(), min(), sum(), avg()

mean_value = mean(nums)
mean_value
```

```python
# Given a list of numbers, only return the values that are greater/less than the mean
```

```python
def new_fun():
    print(f"Below the mean: {list(filter(lambda x: x < mean_value, nums))}")
    print(f"Above the mean: {list(filter(lambda x: x > mean_value, nums))}")
    
new_fun()
```

```python
from functools import reduce

my_list = [1, 2, 3, 7, 10]

def my_func(x, y):
    return x*y

total = 0
for n in my_list:
    total+=n
print(total)
```

```python
reduce(my_func, my_list)
```

```python
# Implementing a base case = the condition for your function to end
def factorial(num):
    #Base Case
    if num < 1:
        return 1
    else:
        print(num)
#         print(num-1)
        return factorial(num-1) * num
    
factorial(5)
```

```python
def countdown(word):
    for char in word:
        print(char, end=' ')
        
countdown("work")[2]
```

```python
def countdown(word):
    for char in word:
        return char
    
countdown("work")[0]
```

```python
def countdown(word):
    for char in word:
        yield char
    
countdown("work")[2]
```

```python
def square_countdown(num):
    while num <= 10:
        yield { num: num**2 }
        num+=1
        
list(square_countdown(2))
```

```python
new_list = []
num = 0
while num <= 10:
    new_list.append({ num: num**2 })
    num+=1
    
new_list
```
